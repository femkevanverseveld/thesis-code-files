```python
# Remove 'FIND_FAILED', 'FIT_FAILED', and 'NA' values
values_to_remove = ["FIND_FAILED", "FIT_FAILED"]

# Create new list to hold the processed DataFrames
processed_dataframes = []

for df in data_frames:
    # Drop NA values
    df.dropna(inplace=True)
    
    # Remove rows containing "FIND_FAILED" or "FIT_FAILED" in any column
    df = df[~df.isin(values_to_remove).any(axis=1)]
    
    processed_dataframes.append(df)

# Convert list back to a tuple
data_frames = tuple(processed_dataframes)
```

```python
# Create list of columns to drop
columns_to_drop = ['Condition', 'Arousal', 'Valence', 'Video Time']

# List comprehension to drop the columns for each DataFrame in the tuple
data_frames = [df.drop(columns=columns_to_drop, errors='ignore') for df in data_frames]
```

```python
# Concatenate DataFrames into a single DataFrame
dataset = pd.concat(data_frames)

# Display DataFrame
dataset
```

```python
# One-Hot Encoding
df_encoded = pd.get_dummies(dataset, columns=['Mouth', 'Left Eye', 'Right Eye', 'Left Eyebrow', 'Right Eyebrow'])

expected_columns = [
    'Mouth_Open', 'Mouth_Closed',
    'Left Eye_Open', 'Left Eye_Closed',
    'Right Eye_Open', 'Right Eye_Closed',
    'Left Eyebrow_Neutral', 'Left Eyebrow_Raised', 'Left Eyebrow_Lowered',
    'Right Eyebrow_Neutral', 'Right Eyebrow_Raised', 'Right Eyebrow_Lowered'
]

for col in expected_columns:
    if col not in df_encoded.columns:
        df_encoded[col] = 0

# Create aggregation functions dynamically for expected_columns and additional columns
additional_cols = ['Neutral', 'Happy', 'Sad', 'Angry', 'Surprised', 'Scared', 'Disgusted']
agg_funcs = {col: ['mean', 'std', 'skew', 'median', lambda x: x.max() - x.min()] for col in (expected_columns + additional_cols)}

# Group by 'ID' and aggregate statistics
df_aggregated = df_encoded.groupby('ID').agg(agg_funcs)

# Custom column naming
df_aggregated.columns = ['_'.join(col).strip().replace('<lambda_0>', 'range') for col in df_aggregated.columns.values]

# Add back 'Artificiality' and 'Expressiveness'
artificiality_expressiveness = dataset[['ID', 'Artificiality', 'Expressiveness']].drop_duplicates().set_index('ID')
df_aggregated = df_aggregated.join(artificiality_expressiveness)

# Reset the index to make 'ID' a column
df_aggregated = df_aggregated.reset_index()

# Display DataFrame
df_aggregated
```

```python
# Filter out rows where 'Artificiality' or 'Expressiveness' is 3
df_aggregated = df_aggregated[(df_aggregated['Artificiality'] != 3) & (df_aggregated['Expressiveness'] != 3)]
```

```python
# Exclude 'ID' from df_aggregated
data2 = df_aggregated.iloc[:, 1:]
```
